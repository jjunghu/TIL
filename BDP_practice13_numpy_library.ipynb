```python
# numpy를 사용한 버블 차트 그리기

import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(10, 100, 200)   # randint(a,b,n) : a이상 b미만인 정수 n개
y = np.random.randint(10, 100, 200)
size = np.random.rand(200) * 100      # rand(n) : 0~1 사이에 있는 n개의 실수

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha = 0.7)
plt.colorbar()
plt.show()
```

```python
# unit 10에서 사용한 버블차트 그리기

import random
import matplotlib.pyplot as plt

x = []
y = []
size = []

for i in range(200) :
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10,100))
    
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# arange() 함수 : 연속된 숫자로 데이터를 생성

import numpy as np
print(np.arange(3))          # 0부터 3미만의 연속된 정수
print(np.arange(3, 7))       # 3이상 7미만의 연속된 정수
print(np.arange(3, 7, 2))    # 3이상 7미만의 연속된 정수 중 2만큼의 간격을 두고 생성
```

```python
# linspace() 함수

# arange() : 특정 간격에 해당하는 값을 생성하고 싶을 때
# linspace() : 특정 개수의 구간으로 나눈 값을 생성하고 싶을 때

import numpy as np
a = np.arange(1, 2, 0.1)   # 1이상 2미만 구간에서 0.1간격으로 실수 생성
b = np.linspace(1, 2, 11) # 1부터 2까지 11개 구간으로 나눈 실수 생성

print(a)
print(b)
```

```python
# numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)    # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50        # x에 저장된 값 중 절댓값이 50보다 큰 값
mask2 = abs(y) > 50        # y에 저장된 값 중 절댓값이 50보다 큰 값

x = x[mask1 + mask2]
y = y[mask1 + mask2]

size = np.random.rand(len(x)) * 100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
