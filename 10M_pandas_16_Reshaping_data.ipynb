### 1. 데이터 재형성_Reshaping Data

- sort_values : 특정 열의 데이터를 기준으로 데이터프레임을 정렬    
- df.rename : 데이터프레임 내의 행, 열의 이름을 변경    
- df.sort_index :Index 별로 정렬    
- df.reset_index : Index가 없을 때, Index를 생성해서 정렬
- df.drop : 특정 컬럼에서 변수를 제거할 때

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df =  sns.load_dataset('mpg')

# 데이터 프레임 형성 확인
df.shape
```

```python
df.head()
```

-  sort_values : 특정 열의 데이터를 기준으로 데이터프레임을 정렬    

```python
df.sort_values('mpg').head()
```

```python
df.sort_values('cylinders').head()
```

```python
df.sort_values('mpg', ascending=False).head()

# ascending = False: 역순
# default는 True
```

- df.rename : 데이터프레임 내의 행, 열의 이름을 변경   

```python
df.rename(columns = {'model_year':'year'}).head()
```

```python
df.head()
```

```python
df = df.rename(columns = {'model_year':'year'})
df.head()
```

- df.sort_index :Index 별로 정렬  

```python
df.sort_index().head()
```

- df.reset_index : Index가 없을 때, Index를 생성해서 정렬

- df.drop : 특정 컬럼에서 변수를 제거할 때

```python
df2 = df.drop(columns=['mpg', 'year'])
df2.head()
```

```python
df3 = df.drop(columns=['mpg', 'year', 'name'])
df3.head()
```
