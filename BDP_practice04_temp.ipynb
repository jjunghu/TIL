```python
# 1월과 8월 데이터를 히스토그램으로 시각화하기 (최고기온)

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_book.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data: 
    if row[-1] != '' :
        month = row[0].split('-')[1]
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))
            
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')

plt.legend()
plt.show()

                    
```

```python
# HW 1월의 최저기온 데이터를 히스토그램으로 시각화

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_book.csv')
data = csv.reader(f)
next(data)

jan = []

for row in data : 
    if row[-2] != '' :
        if row[0].split('-')[1] == '01' :
            jan.append(float(row[-2]))
            
plt.hist(jan, bins = 100, color = 'skyblue')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('1월의 최저기온')
plt.show()
```

```python
## 상자그림(boxplot)

import matplotlib.pyplot as plt
import random

result = []

for _ in range(13) : 
    result.append(random.randint(1,1000))

print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
# 최고기온 데이터를 월별로 구분하여 표현

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_book.csv')
data = csv.reader(f)
next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data : 
    if row[-1] != '' :
        month[int(row[0].split('-')[1]) -1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
# 8월 일별 기온데이터를 boxplot으로 표현 (최고기온)

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_book.csv')
data = csv.reader(f)
next(data)

day = []
for _ in range(31):
    day.append([])

for row in data : 
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :
                day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))
                

plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi = 300)
            
plt.boxplot(day, showfliers = False)  # outliers 출력 안함
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('8월 일별 최고기온 데이터')

plt.show()
```
