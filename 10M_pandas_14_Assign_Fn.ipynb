assign() : 새로운 컬럼을 데이터프레임에 배정(Assign)하는 함수

### assign으로 새로운 컬럼 만들기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A':range(1,11), 'B':np.random.randn(10)})
df
```

```python
df.assign(ln_A = lambda x: np.log(x.A))
```

```python
df.assign(ln_A = lambda x: np.log(x.A)).head()
```

```python
df['ln_A'] = np.log(df.A)
df['ln_A']
```
