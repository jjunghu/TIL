* drop_duplicates(): 중복된 값 삭제

```python
df
```

```python
df.drop_duplicates()
# 현재 중복된 값 없음
```

```python
# 임의로 중복값 생성

df2 = pd.DataFrame(
        {'a' : [4, 5, 6, 6],
        'b' : [7, 8, 9, 9],
        'c' : [10, 11, 12, 12]},
        index = pd.MultiIndex.from_tuples(
        [('d',1), ('d',2), ('e',2), ('e',3)],
        names=['n','v']))

df2
```

```python
df2.drop_duplicates()
```

```python
df2
# 여전히 중복된 값이 남아있음
```

```python
df2.drop_duplicates(inplace=True)
# inplace of noun: ~을 대신하다
# python에서 inplace 사용 권장하지 않음

df2
```

```python
df2 = df2.drop_duplicates()
df2
```

```python
df2 = pd.DataFrame(
        {'a' : [4, 5, 6, 6],
        'b' : [7, 8, 9, 9],
        'c' : [10, 11, 12, 12]},
        index = pd.MultiIndex.from_tuples(
        [('d',1), ('d',2), ('e',2), ('e',3)],
        names=['n','v']))

df2
```

```python
df2 = df2.drop_duplicates(keep='last')
df2

# 마지막행에 있는 값을 사용함
```
