```python
# 그래프에 제목 넣기

import matplotlib.pyplot as plt
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# 그래프에 범례 넣기

import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend()
plt.show()
```

```python
# 범례 위치 변경

import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend(loc = 9)
plt.show()
```

```python
# 그래프 색상 바꾸기

import matplotlib.pyplot as plt
plt.title('color')

plt.plot([10, 20, 30, 40], color = 'hotpink', label = 'hotpink')
plt.plot([40, 30, 20, 10], color = 'skyblue', label = 'skyblue')

plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기

import matplotlib.pyplot as plt
plt.title('linestyle')

plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], color = 'g', linestyle = ':', label = 'dotted')

plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기 간단하게!

import matplotlib.pyplot as plt
plt.title('linestyle2')

plt.plot([10, 20, 30, 40], 'r--')
plt.plot([40, 30, 20, 10], 'g:')

plt.show()
```

```python
# 마커 모양 바꾸기

import matplotlib.pyplot as plt
plt.title('marker')

plt.plot([10, 20, 30, 40], 'r.', label = 'circle')
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up')

plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기 2

import matplotlib.pyplot as plt
plt.title('marker2')

plt.plot([10, 20, 30, 40], 'r.--', label = 'circle')
plt.plot([40, 30, 20, 10], 'g^:', label = 'triangle up')

plt.legend()
plt.show()
```
