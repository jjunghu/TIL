```python
import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/age.csv')
data = csv.reader(f)

result = []
name = input('인구 구조가 알고싶은 지역의 이름(읍면동 단위)을 입력하세요: ')

for row in data : 
    if name in row[0] : 
        for i in row[3:] :
            result.append(int(i.replace(',','')))
            
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
# 신도림의 인구구조를 막대그래프로 표현

import csv
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/age.csv')
data = csv.reader(f)

result = []

for row in data : 
    if '신도림' in row[0] :
        for i in row[3:] : 
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
# 막대그래프를 수평방향으로 barh()

plt.barh(range(101), result)
plt.show()
```

```python
# 신도림의 남녀성별 데이터 그래프로 표현

import csv
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data : 
    if '신도림' in row[0] : 
        for i in range(0,101) : 
            m.append(-int(row[i + 3]))
            f.append(int(row[-i -1]))
f.reverse()

import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
# 신도림의 남녀성별 데이터 그래프로 표현 2

import csv
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data : 
    if '신도림' in row[0] : 
        for i in row[3 : 104] :
            m.append(-int(i))
        for i in row[106:207] : 
            f.append(int(i))
            
import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
# 내가 사는 지역의 남녀 성별 인구분포를 항아리 모양으로 나타내기

import csv
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data : 
    if '온양5동' in row[0] : 
        for i in range(0,101) :
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('온양5동 지역의 남녀 성별 인구분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
# 우리동네 인구구조를 항아리모양 그래프로 그리기
# 원하는 지역의 이름을 넣을 수 있는

import csv
import matplotlib.pyplot as plt

f = open('C:/Users/jjunghu/Desktop/jjunghu/data/gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('찿고싶은 지역의 이름(읍면동 단위)을 알려주세요 : ')

for row in data : 
    if name in row[0] : 
        for i in range(0,101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()

plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
