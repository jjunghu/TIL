```python
import numpy as np
import pandas as pd
```

```python
df3 = pd.DataFrame(
        {'a' : [4, 5, 6, 6, np.nan],
        'b' : [7, 8, np.nan, 9, 9],
        'c' : [10, 11, 12, np.nan, 12]},
        index = pd.MultiIndex.from_tuples(
        [('d',1),('d',2),('e',2),('e',3),('e',4)],
        names = ['n','v']))

df3
```

* df3.nlargest[n, 'value']

```python
df4 = pd.DataFrame({'a' : [1, 10, 8, 11, -1],
                   'b' : list('abcde'),
                   'c' : [1.0, 2.0, np.nan, 3.0, 4.0]})

df4
```

```python
df4.nlargest(1, 'a')    # a컬럼에서 가장 큰 값만 가지고 오기
```

```python
# df4.nlargest(1, 'b')
# Type Error 타입에러 발생, 숫자 컬럼에서만 가져올 수 있음
```
