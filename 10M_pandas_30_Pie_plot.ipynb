### Pie plot
- Seaborn 등 다른 모듈에서는 파이차트 지원하지 않음
- 면적으로 나타나 오해의 소지가 존재하기 떄문
- DataFrame 뒤에 plot.pie() 함수를 호출하면 구현됨

```python
import pandas as pd
import numpy as np
import matplotlib as mpl

# Jupyter Notebook에 그래프 표시를 위해
%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
series = pd.Series(np.random.rand(4), 
                   index=['a', 'b', 'c', 'd'], name='series')
series
```

```python
series = pd.Series(3 * np.random.rand(4),
                  index = ['a', 'b', 'c', 'd'], name = 'series')
series
```

```python
series.plot.pie()
```

```python
series.plot.pie(figsize=(6,6))
```

```python
series.plot.pie(labels=['AA', 'BB', 'CC', 'DD'],
               colors = ['r', 'g', 'b', 'c'],
               autopct = '%.2f',
               fontsize=20,
               figsize=(6,6))
```

```python
df = pd.DataFrame({'mass':[0.330, 4.87, 5.97],
                 'radius':[2439.7, 6051.8, 6378.1]}, 
                 index=['Marcury', 'Venus', 'Earth'])

df
```

```python
df_m = df.mass    # df['mass']
plot = df_m.plot.pie(figsize=(10,5))
```

```python
plot = df.plot.pie(subplots = True, figsize=(10,5))
```
