```python
# 8월의 최고기온 데이터를 그래프로

import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_new.csv')
data = csv.reader(f)

next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
            
plt.plot(result, 'hotpink')
plt.show()
```

```python
# 2월 14알 최고기온 그래프

import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_new.csv')
data = csv.reader(f)

next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))
            
plt.plot(result, 'hotpink')
plt.show()
        
```

```python
# 1983년 이후 2월 14일 최고기온/최저기온 그래프

import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_new.csv')
data = csv.reader(f)

next(data)
high = []
low = []

for row in data:
    if row[-1] != '':
        if 1983 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.show()
```

```python
# 제목 넣기

import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_new.csv')
data = csv.reader(f)

next(data)
high = []
low = []

for row in data:
    if row[-1] != '':
        if 1983 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('2월14일의 기온변화 그래프')
plt.show()
```

```python
# 내 생일의 기온 변화 그래프

import csv
import matplotlib.pyplot as plt
f = open('C:/Users/jjunghu/Desktop/jjunghu/data/seoul_new.csv')
data = csv.reader(f)

next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1994 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '16':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.plot(high, 'red', label = 'high')
plt.plot(low, 'blue', label = 'low')

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('내 생일의 기온 변화 그래프')

plt.legend()
plt.show()
```
